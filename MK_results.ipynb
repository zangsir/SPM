```python
#see original csv file and if they match with altered file

def read_csv_MK_file(csv_file):
    #numbers only each line
    f=open(csv_file,'r').read().split('\n')
    total=[]
    for k in range(len(f)):
        line=f[k]
        #print k
        if line!='':
            l=line.split(',')

            l=[float(i) for i in l]
            total.append(l)
    return total


def read_csv_ori_file(csv_file):
    #each line has numbers and string attributes
    f=open(csv_file,'r').read().split('\n')
    total=[]
    for k in range(len(f)):
        line=f[k]
        #print k
        if line!='':
            l=line.split(',')
            #attr=-1*num_nominal
            #numbers=l[:attr]
            #l=[float(i) for i in numbers]
            total.append(l)
    return total
```
