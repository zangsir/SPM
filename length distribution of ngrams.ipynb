# look at the distribution of length of syllables in number of samples

```python

def read_ori_file(whole_file):
    lines=open(whole_file,'r').read().split('\n')
    all_syl_len=[]
    for line in lines:
        l=line.split(',')[:-4]
        all_syl_len.append(len(l))
    return all_syl_len
```

```python
%matplotlib inline
import pylab as plt
import numpy as np

def plot_hist(all_syl_len):
    plt.hist(all_syl_len,bins=30)
    plt.title('histogram of the length of syllables in number of samples' )
    print np.std(all_syl_len)
```

```python
#unigram
whole_file='syl_norm_whole_meta.csv'

all_syl_len=read_ori_file(whole_file)
print len(all_syl_len)
```

```python
plot_hist(all_syl_len)
```

It seems like the length of one syllable has a mean of around 150-200 samples with a std of 63 samples. Therefore, if we have motif length of 450 samples, for example, it roughtly means we're looking for trigrams motifs.
