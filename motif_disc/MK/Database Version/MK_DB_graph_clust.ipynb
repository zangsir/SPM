# doing conversion of tuple file into tuple_graph file

the format is ts1, ts2, weight(dist)

```python
from os import listdir
path='mk_txt'
onlyfiles = [ f for f in listdir(path) if f.endswith("_10_tuple.txt")]
print onlyfiles
for filename in onlyfiles:
    inputfile=path +'/'+filename
    output=inputfile.split('.')[0]+'_graph.txt'
    f=open(inputfile,'r').read().split('\n')
    g=open(output,'w').close()
    g=open(output,'a')
    for line in f[1:]:
        if line!='':
            l=line.split(',')
            newline=[l[1],l[2],l[0]]
            g.write(' '.join(newline)+'\n')
    g.close()
            
```

```python
in_degrees= g.degree() # dictionary node:degree
in_values= sorted(set(in_degrees.values()))
in_hist= [in_degrees.values().count(x) for x in in_values]
```

```python
plt.plot(in_values, in_hist, 'ro-') # in-degree
plt.figure() # you need to first do 'import pylab as plt'
plt.grid(True)
plt.loglog(in_values, in_hist, 'ro-') # in-degree

plt.xlabel('Degree')
plt.ylabel('Number of nodes')
plt.title('unigram 100p MK_DB motifs(loglog)')
```

```python
# print out sorted nodes by their degree
for i in sorted(in_degrees,key=in_degrees.get,reverse=True):
    print i,in_degrees[i]#using the dictionary to print out the value, i.e., the degree number
```

```python

def get_con_index(tup,target):
    """ in tup (10,13374) returns the index that's not target. target is 13374, then return 0 """
    if len(tup)!=2:
        raise ValueError('tuple length is not 2')
    if tup.index(target)==1:
        return 0
    elif tup.index(target)==0:
        return 1
    else:
        raise ValueError('target not in tuple or tuple length is not 2')
```

```python
get_con_index((13374, 89),13374)
```

```python
#among the top degreed, what are the motif pairs (nodes) they are connected 

print "central node: connected nodes"
cluster_dict={}#key is the central node, value is the nodes it is connected to
all_edges=g.edges()
sort_nodes=sorted(in_degrees,key=in_degrees.get,reverse=True)
k=15 #get top k most connected motifs
print 'k=',k
for i in range(k):
    node=sort_nodes[i]
    
    connected_nodes=[j[get_con_index(j,node)] for j in all_edges if node in j]
    print node,":",connected_nodes 
    cluster_dict[node]=connected_nodes
    
```
